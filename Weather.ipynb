```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv("results.csv")
df.head()
```

```python
df.columns.unique()
```

```python
df.columns = ['Year', 'NewDelhi', 'Global']
```

```python
df.info()
```

```python
df.describe()
```

```python
df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)
df['NewDelhi'] = df.rolling(window=10)['NewDelhi'].mean()
df['Global'] = df.rolling(window=10)['Global'].mean()
```

```python
params = {
    'legend.fontsize': 24,
    'figure.figsize': (18, 10),
    'axes.labelsize': 20,
    'axes.titlesize': 20,
    'xtick.labelsize': 'x-large',
    'ytick.labelsize': 'x-large'
}
plt.rcParams.update(params)

plt.title('\nLine Chart of New Delhi vs. Global Temperature\n(10-year moving average)')
plt.ylabel('Temperature (in Degree Celsius)')
plt.xlabel('Year')
plt.xticks(np.arange(1800, 2013, step=10))
plt.yticks(np.arange(7, 30, step=1))

plt.plot(df.Year, df.NewDelhi, linewidth=4, color='Red', ls = 'dashdot', label='New Delhi')
plt.plot(df.Year, df.Global, linewidth=4, color='green', ls = 'dashdot', label='Global')
plt.grid(True)

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

plt.show()
```
